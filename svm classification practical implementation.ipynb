```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df.head(4)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# In this dataset we can see that userid and gender is not that useful so we will focus on other two
x=df.iloc[:,2:4]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# feature Scaling

from sklearn.preprocessing import StandardScaler
ssc=StandardScaler()
x_train=ssc.fit_transform(x_train)
x_test=ssc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
# checking accuracy and classification report
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy of the model is:- ",accuracy_score(y_test,y_pred)*100)

print(classification_report(y_pred,y_test))
```

```python
# Will take some different hyperparameters now
svc1=SVC(kernel='rbf',gamma=10,C=5)
svc1.fit(x_train,y_train)
```

```python
y_pred1=svc1.predict(x_test)
y_pred1
```

```python
print(accuracy_score(y_pred1,y_test))
```

```python
print(classification_report(y_pred1,y_test))
```

```python
# lets just check with polynomial kernel
svc2=SVC(kernel='poly',degree=4,gamma=10,C=5)
svc2.fit(x_train,y_train)
y_pred2=svc2.predict(x_test)
y_pred2
```

```python
print(accuracy_score(y_pred2,y_test))
```

```python
print(classification_report(y_pred2,y_test))
```

```python
svc3=SVC(kernel='poly',degree=2,gamma=15,C=7)
svc3.fit(x_train,y_train)
y_pred3=svc3.predict(x_test)
y_pred3
```

```python
print(accuracy_score(y_pred3,y_test))
```

```python
svc4=SVC(kernel='poly',degree=3)
svc4.fit(x_train,y_train)
y_pred4=svc4.predict(x_test)
y_pred4
```

```python
print(accuracy_score(y_pred4,y_test))
```

```python
print(classification_report(y_pred4,y_test))
```
