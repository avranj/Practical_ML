```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=32)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
# lets do some feature scaling on training and testing  dataset
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
print(sc.fit(x_train) )

sc.fit(x_test)
```

```python
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train_sc=pd.DataFrame(x_train,columns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'])
```

```python
x_test_sc=pd.DataFrame(x_test,columns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'])
```

```python
x_train_sc
```

```python
x_test_sc
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
models=[LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
for model in models:
    model.fit(x_train,y_train)
    prediction_set=model.predict(x_test)
    acc=accuracy_score(y_test,prediction_set)
    print("accuracy score of ",model,' = ',acc)
```

```python
# lets try to scale our independent features directly
sc.fit(x)
```

```python
x_scaled=sc.transform(x)
x_scaled
```

```python
x_scaled_sc=pd.DataFrame(x_scaled,columns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'])
```

```python
x_scaled_sc
```

```python
# again applying models and train test split
X_train,X_test,Y_train,Y_test=train_test_split(x_scaled_sc,y,test_size=0.3,random_state=32)
```

```python
# again fitting in the models with same parameters
for model in models:
    model.fit(X_train,Y_train)
    prediction_set=model.predict(X_test)
    acc=accuracy_score(Y_test,prediction_set)
    print("accuracy score of ",model,' = ',acc)
```

# observation till now feature scaling enhanced the model performance

## Now crossValidation approach 

Logistic regression

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_lr=cross_val_score(LogisticRegression(max_iter=1000),x_scaled_sc,y,cv=5)
print(cv_lr)
```

```python
mean_acc= sum(cv_lr)/len(cv_lr)
mean_acc
```

SVC

```python
cv_lr=cross_val_score(SVC(kernel='linear'),x_scaled_sc,y,cv=5)
print(cv_lr)
mean_acc= sum(cv_lr)/len(cv_lr)
print(mean_acc)
```

From here we can understand the average score is decreasing that means our model was overfittting on some datasets

```python
models_cv = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
for model in models_cv:
    cv_score=cross_val_score(model,x_scaled_sc,y,cv=5)
    mean_score=sum(cv_score)/len(cv_score)
    mean_score=mean_score*100
    mean_score=round(mean_score,2)
    print("cross validation score of  ",model,' = ',mean_score)
    
    
```
